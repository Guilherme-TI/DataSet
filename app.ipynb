A análise abaixo foi feita utilizando dados salariais com base no país e na raça que foram disponibilizados no site www.kaggle.com
Perguntas que tentaremos responder com a análise:
- Existe uma disparidade salarial entre homens e mulheres?
- Existe uma disparidade salarial entre as diferentes raças?

```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
#Visualização da base de dados
tabela = pd.read_csv('Salario.csv')
display(tabela)

plt.style.use('ggplot')
```

```python
#Dados do faturamento total
faturamento_total= tabela['Salary'].sum()
display(faturamento_total)
```

```python
#Dados do faturamento por gênero
faturamento_por_genero = tabela[['Gender', 'Salary']].groupby('Gender').sum()
display(faturamento_por_genero)
```

```python
_ = tabela['Race'].value_counts().plot.bar()
_ = plt.title('Histograma de Raça')
```

```python
_ = plt.hist(tabela['Age'])
_ = plt.title('Histograma de idade')
```

```python
_ = tabela['Gender'].value_counts().plot.bar()
_ = plt.title('Histograma de gênero')
```

```python
_ = plt.hist(tabela['Salary'])
_ = plt.title('Histograma de Salario')
```
